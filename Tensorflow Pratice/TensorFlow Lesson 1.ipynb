```python
from tensorflow import keras
import tensorflow as tf
```

```python
fashion_mnist = keras.datasets.fashion_mnist
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
class stop(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('loss')<0.25):
            print('\nLoss is low, cancelling training!')
            self.model.stop_training = True
callbacks = stop()
```

```python
model = Sequential([
    Flatten(),
    Dense(512, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=50, callbacks=[callbacks])
```

```python
model.evaluate(x_test, y_test)
```

```python
model3 = Sequential([
    Conv2D(16, (3,3), activation='relu', input_shape=(300,300,3)),
    MaxPooling2D(2,2),
    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model3.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop

model3.compile(optimizer=RMSprop(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
import os
os.getcwd()
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1/255)
train_generator = train_datagen.flow_from_directory(
    'horse-or-human',
    target_size=(300,300),
    batch_size=128,
    class_mode='binary'
)
```

```python
model3.fit(train_generator,
          steps_per_epoch = 8,
          verbose = 2,
          epochs = 15
)
```

```python
testdir = os.getcwd()+'/testmode'
test = os.listdir(testdir)
```

```python
from IPython.display import Image
```

```python
from keras.preprocessing import image
for loc in test:
    path = testdir + '/' + loc
    img = image.load_img(path, target_size=(300,300))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    images = np.vstack([x])
    classes = model3.predict(images, batch_size=10)
    print(classes[0])
    if classes[0] > 0.5:
        print(loc + " is a human")
        display(Image(filename=path))
    else:
        print(loc + " is a horse")
        display(Image(filename=path))
```
